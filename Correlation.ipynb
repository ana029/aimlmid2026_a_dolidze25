# Pearson Correlation Analysis

The data points were extracted manually from the online graph at the following address:

**http://max.ge/aiml_midterm/51892_html**

The coordinates of the blue points were obtained by hovering over each point on the graph.
The goal of this task is to compute Pearson's correlation coefficient and visualize the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Extracted data points from the online graph
x = np.array([-9, -7, -5, -3, -1, 1, 3, 5, 7, 9])
y = np.array([7.9, 6.7, 4.8, 5.1, 0.9, -0.5, -2.8, -4.7, -6.5, -8.3])
```

```python
# Calculate Pearson correlation coefficient
pearson_corr = np.corrcoef(x, y)[0, 1]

pearson_corr
```

```python
plt.figure()
plt.scatter(x, y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Scatter Plot of Extracted Data Points")
plt.show()
```

```python
# Save the scatter plot as an image
plt.figure()
plt.scatter(x, y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Scatter Plot of Extracted Data Points")
plt.savefig("correlation_plot.png")  # PNG
plt.show()
```
